```python
# Crear una clase
class circle:
    
    def __init__(self, radius, color):
        self.radius = radius # atributos
        self.color = color   # atributos
        
    def area(self):
        return math.pi * (self.radius ** 2)
```

```python
c = circle(5, "blue")
```

```python
c.radius
```

```python
c.color
```

```python
print(dir(c))
```

```python
c.area()
```
