```python
x = np.arange(-10, 10, 0.25)
```

```python
x.shape
```

```python
y = np.arange(-10, 10, 0.25)
```

```python
y.shape
```

```python
X, Y = np.meshgrid(x, y)
```

```python
X.shape
```

```python
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))
```

```python
Z.shape
```

```python
Z
```

```python
fig = plt.figure(figsize = (10, 8))
ax = plt.axes(projection = "3d")
ax.scatter3D(X, Y, Z)
plt.show()
```

```python
fig = plt.figure(figsize = (10, 8))
ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z)
plt.show()
```
