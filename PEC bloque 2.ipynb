<div style="background:#5D6D7E;padding:20px;color:#ffffff">
Escribe una funcin que reciba como argumento un dataframe Pandas, obtenga las 5 columnas con mayor nmero de valores nulos y devuelva una serie Pandas conteniendo los nombres de dichas columnas como ndices y el nmero de nulos como valores. Aplcala al dataframe Titanic descargado de la librera Seaborn.
</div>

```python
def col5mas(dataframe):
    dict_df = dataframe.isnull().sum().to_dict()
    
    df = pd.DataFrame(data = dict_df.values(), 
                  index = dict_df.keys(), 
                  columns = ["Total de NaN"]).sort_values(by = "Total de NaN", 
                                                          ascending = False)
    
    return df.iloc[0:5,:]

titanic = sns.load_dataset("titanic")
col5mas(titanic)
```

<div style="background:#5D6D7E;padding:20px;color:#ffffff">
Escribe una funcin que reciba como argumento un dataframe Pandas y sustituya los valores nulos de cada columna por el valor ms frecuente en la columna. Aplcala al dataframe Titanic descargado de la librera Seaborn.
</div>

```python
# titanic.mode() - valores ms frecuentes
# titanic.mode().loc[0] - usar la primera linea del nuevo DF(titanic.mode())
# titanic.mode().loc[0].to_dict() # transforma el DF en un dict

def sus_freq(df):
    df.fillna(value = df.mode().loc[0].to_dict(), inplace = True)
    return df.isnull().sum() # o return df.head()

sus_freq(titanic)
```
